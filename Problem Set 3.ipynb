```python
import matplotlib.pyplot as pp
import pandas as pd
import numpy as np

marry=pd.read_csv("us-marriages-divorces-1867-2014.csv")
#marryM=pd.read_csv("us-marriages-divorces-1867-2014.csv", usecols=[1])
#marryD=pd.read_csv("us-marriages-divorces-1867-2014.csv", usecols=[2])



nMarryYear = np.array(marry['Year'])
nMarryM = np.array(marry['Marriages']/marry['Population'])
nMarryD = np.array(marry['Divorces']/marry['Population'])

pp.plot(nMarryM,nMarryYear)
pp.plot(nMarryD,nMarryYear)
pp.legend(['Marriage per Capita','Divorce per Capita'])
pp.xlabel('Marriages / Divorces per Capita')
pp.ylabel('Year')
pp.title('Marriages / Divorces per Capita for Each Year')
```

# Question 7

```python
import matplotlib.pyplot as pp
import pandas as pd
import numpy as np

marry=pd.read_csv("us-marriages-divorces-1867-2014.csv")
marryYear = marry[marry['Year'].isin([1900,1950,2000])].copy()

marryYear['MarriageperCapita'] = marryYear['Marriages']/marryYear['Population']
marryYear['DivorceperCapita'] = marryYear['Divorces']/marryYear['Population']

marryYear2 = marryYear[['Year','MarriageperCapita','DivorceperCapita']]

marryYear2 = marryYear2.set_index(['Year'])
#xyz = np.arange(len(nMarryYear))

marryYear2.plot(kind='bar')

pp.legend(['Marriage per Capita','Divorce per Capita'])
pp.ylabel('Marriages / Divorces per Capita')
pp.xlabel('Year')
pp.title('Marriages / Divorces per Capita in 1900, 1950 and 2000')
```
